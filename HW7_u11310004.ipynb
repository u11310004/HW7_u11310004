```python
import math

# 原始數據
lat_deg = 48.1195
h_land = 487.9
g_obs = 980717.39  # mGal

lat = math.radians(lat_deg)
sin2 = math.sin(lat)**2
sin4 = sin2**2

# 理論正常重力 (1967公式)
gamma0 = 978032.68 * (1 + 0.0053024*sin2 - 0.0000058*sin4)

print("=== a) 陸地測站 (海拔 487.9 m) ===")
FAC_land = -0.3086 * h_land
BC_land = -0.1119 * h_land
FAA_land = g_obs - gamma0 - FAC_land
BA_land = FAA_land - BC_land

print(f"理論重力 γ₀       = {gamma0:.3f} mGal")
print(f"自由空氣校正      = {FAC_land:+.3f} mGal")
print(f"布格校正          = {BC_land:+.3f} mGal")
print(f"自由空氣異常 FAA  = {FAA_land:.3f} mGal")
print(f"布格異常 BA       = {BA_land:.3f} mGal")

print("\n=== b) 海洋測站 (水深 487.9 m) ===")
h_ocean = -487.9
FAC_ocean = -0.3086 * h_ocean
BC_ocean = -0.0689 * abs(h_ocean)   # 海水替換校正
FAA_ocean = g_obs - gamma0 - FAC_ocean
BA_ocean = FAA_ocean - BC_ocean

print(f"自由空氣校正      = {FAC_ocean:+.3f} mGal")
print(f"布格校正 (海洋)   = {BC_ocean:+.3f} mGal")
print(f"自由空氣異常 FAA  = {FAA_ocean:.3f} mGal")
print(f"布格異常 BA       = {BA_ocean:.3f} mGal")
```
